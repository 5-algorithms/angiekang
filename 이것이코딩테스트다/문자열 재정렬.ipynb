```python
s = input()

digit = 0
alpha = []
# 하나씩 돌아가며 검사, 숫자는 더해주고 문자는 따로 저장해주자
for data in s:
    if data.isalpha():
        alpha.append(data)
    if data.isdigit():
        digit+=int(data)
        
# 문자열 정렬
alpha.sort()
# 문자열  + 숫자
digit = str(digit)
alpha.append(digit)
result = ''.join(alpha)
print(result)
```
