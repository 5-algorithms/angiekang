```python
s = input()

digit = 0
alpha = []
#   ,     
for data in s:
    if data.isalpha():
        alpha.append(data)
    if data.isdigit():
        digit+=int(data)
        
#  
alpha.sort()
#   + 
digit = str(digit)
alpha.append(digit)
result = ''.join(alpha)
print(result)
```
